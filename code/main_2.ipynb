# Analysing the datasets

## For Train dataset

* Unique id column which had info (catoegory ,cat num and store number)
* item id which is in price and calendar 
* dep_id denotes which department the product is
* store_id denotes which store the product is
* and number of d_... columns from 1 to 1941 days

## price dataset

* store_id same as in train
* item_id same as in train
* wm_yr_wk (wallmart_year_week) this gives the chronological sequence but not ordered
* sell_price gives the price of the product that week

## calendar dataset

* date gives the date of the record
* wm_yr_wk same as in price dataset but it is ordered in chronlogical order
* weekday and wday gives the same thing but we can say wday is label encoded
* month gives what month ranges(1 to 12)
* year gives what year it is
  * These last 3 or 4 columns can be derived from date

* d gives the details about the count of day which is same number in columns of train ranges(1 to 1941)
* event_name and event_type (4 columns) these gives the event and type of event happening in that particular day if there are 2 events in a day thus there are 2 columns of each
* snap is a system in US (Supplemental Nutrition Assistance Program) which the government give money to buy food the elegibility is given to certain stores at certain times in (0,1)


## Additional details

* wm_yr_wk has values from 11101 (1_11_01 stands as wm_yr_wk), since in a year there can be 52 week max the last 2 digits cannot be more than 52
* the id in train is combination of item_id and store_id
* the dep_id in train can be derived from dept_id
